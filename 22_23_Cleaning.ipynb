```python
#now for players that switched teams we want to only keep the column with total
```

```python
traded_players = []

for index, row in combined.iterrows():
  if combined.at[index,'Tm'] == 'TOT':
    traded_players.append(combined.at[index,'Player'])
```

```python
#traded_players
```

```python
for index, row in combined.iterrows():
  #if the player was traded and the row does not have team = TOT (the players total season stats), we delete the row
  if combined.at[index,'Player'] in traded_players and combined.at[index,'Tm'] != 'TOT':
    combined.drop(labels = index, axis = 0, inplace = True)
combined = combined.reset_index(drop = 'True')
```

```python
for index, row in salaries.iterrows():
  #if the salary is a string (meaning it isn't null), remove $ and commas and caste to float
  if isinstance(salaries.at[index, 'Salary 2022-23'], str) == True:
    instance = salaries.at[index, 'Salary 2022-23']
    #strip the $ and commas from the provided salary
    instance = instance.replace("$", "")
    instance = instance.replace(",", "")
    #convert the salary to a float and divide by 1,000,000 so that the column holds
    #salary in millions
    salaries.at[index, 'Salary 2022-23'] = (float(instance))/1000000
```

```python
#drop the Rk column from salaries
salaries.drop(columns = ['Rk'], axis = 1, inplace = True)
```

```python
#29 players are missing salaries
salaries['Salary 2022-23'].isna().sum()
```

# Now, we just need to add on 2022-2023 season salary onto each entry

```python
#create a dictionary holding a map from player name to salary
salaryDict = dict(zip(salaries['Player'], (salaries['Salary 2022-23'])))
```

```python
combined.columns
```

# We are able to fill a large portion of the salaries using the salaryDict dictionary. However, around 20 players still need to be assigned a salary

```python
combined['Salary'] = ''
for index, row in combined.iterrows():
  if combined.at[index,'Player'] in salaryDict.keys():
    combined.at[index, 'Salary'] = salaryDict[combined.at[index, 'Player']]
```

```python
#there are only 20 null values that need to be filled in
combined['Salary'].isna().sum()
```
