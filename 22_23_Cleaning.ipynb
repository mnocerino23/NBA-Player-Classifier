```python
salaries = pd.read_csv('/content/drive/MyDrive/Data_Science_Projects/NBA/2022-2023Salaries.csv')
perGame = pd.read_csv('/content/drive/MyDrive/Data_Science_Projects/NBA/perGame2223.csv')
advancedStats = pd.read_csv('/content/drive/MyDrive/Data_Science_Projects/NBA/advanced2223.csv')
```

```python
#drop empty columns from the 2022-2023 salaries dataset
salaries.drop(columns = ["Column1", "Column2", "Column3", "Column4", "Column5", "Column6"], inplace = True)
```

```python
#drop repeated columns and change name of games and player columns so that we can double check our join
advancedStats.drop(['Rk', 'Age' ,'Pos', 'Tm', 'MP'], axis = 1, inplace = True)
advancedStats.rename(columns={'G': 'Games'}, inplace = True)
advancedStats.rename(columns={'Player': 'Name'}, inplace = True)
```

```python
perGame.shape
```

```python
#now join together the advancedStats (use an inner join)
combined = pd.concat([perGame, advancedStats], axis=1)
```

```python
combined.shape
```

```python
combined.head(15)
```

```python
combined.columns
```

```python
#double check that all entries line up from concat
for index, row in combined.iterrows():
  print(combined['Player'][index], combined['Name'][index])
  print(combined['G'][index], combined['Games'][index])
```

```python
combined.drop(['Name', 'Player-additional', 'Rk', 'Unnamed: 19', 'Unnamed: 24'], axis = 1, inplace = True)
```

```python
combined.head()
```

```python
#now for players that switched teams we want to only keep the column with total
```

```python
traded_players = []

for index, row in combined.iterrows():
  if combined.at[index,'Tm'] == 'TOT':
    traded_players.append(combined.at[index,'Player'])
```

```python
traded_players
```

```python
for index, row in combined.iterrows():
  #if the player was traded and the row does not have team = TOT (the players total season stats), we delete the row
  if combined.at[index,'Player'] in traded_players and combined.at[index,'Tm'] != 'TOT':
    combined.drop(labels = index, axis = 0, inplace = True)
combined = combined.reset_index(drop = 'True')
```

# Now, we just need to add on 2022-2023 season salary onto each entry

```python
#cast salaries to float and then add a new column with salary in millions using a lambda function
salaries = salaries.astype({'2022/23':'float'})
salaries = salaries.assign(Salary = lambda x: (x['2022/23']*(1/1000000)))
```

```python
salaries.drop(columns = ['2022/23'], inplace = True)
salaries
```

```python
#create a dictionary holding a map from player name to salary
salaryDict = dict(zip(salaries['Player'], (salaries['Salary'])))
```

```python
salaryDict
```

```python
combined.columns
```

```python
combined['Salary'] = ''
for index, row in combined.iterrows():
  if combined.at[index,'Player'] in salaryDict.keys():
    combined.at[index, 'Salary'] = salaryDict[combined.at[index, 'Player']]
```
