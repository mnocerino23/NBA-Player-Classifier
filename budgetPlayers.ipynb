```python
complete = pd.read_csv('/content/drive/MyDrive/Data_Science_Projects/NBA/endSeasonCombined2022-2023.csv')
```

```python
#limit our dataset to those who have played over 15 games this season
complete = complete.loc[complete['G'] > 15]
```

# Who are the best bang for your buck players?

```python
#add features price/offensive_win_share, price/defensive_win_share, price/win_share
#using lambda functions
complete = complete.assign(S_OWS =lambda x: (x['Salary']/x['OWS']))
complete = complete.assign(S_DWS =lambda x: (x['Salary']/x['DWS']))
complete = complete.assign(S_WS =lambda x: (x['Salary']/x['WS']))
```

```python
complete.head(5)
```

# Best budget offensive players right now:

```python
nonnegative_offense = complete.loc[(complete['OWS'] > 0) & (complete['Age'] >25) & (complete['PER'] >15)]
nonnegative_offense.nsmallest(10, ['S_OWS'])
```

#  Best budget defenders right now:

```python
nonnegative_defense = complete.loc[(complete['DWS'] > 2.5) & (complete['Age'] >25)]
nonnegative_defense.nsmallest(10, ['S_DWS'])
```

# Best budget players

```python
nonnegative_ws = complete.loc[(complete['WS'] > 2) & (complete['Age'] >25)]
nonnegative_ws.nsmallest(15, ['S_WS'])
```

Best budget three & D players:

```python
three_and_d = complete.loc[(complete['3P%'] > 0.37) & (complete['Age'] >25) & (complete['DWS'] >2) & (complete['Salary'] < 13)]
three_and_d.nlargest(10, ['WS'])
```

```python
budget = complete.loc[(complete['Age'] > 24) & (complete['WS'] > 5) & (complete['Salary'] < 13)].sort_values(by = ['WS'], ascending = False)
budget
```

# Introduce a new column storing win shares per game played called 'WSperG'

```python
complete = complete.assign(WSperG =lambda x: (x['WS']/x['G']))
```

# Display the top 20 players in the NBA by Win shares per game played

```python
top_20 = complete[['Player', 'Age', 'USG%', 'TS%', 'PER', 'WSperG', 'Salary']].sort_values(by = ['WSperG'], ascending = False)
top_20.reset_index(inplace = True, drop = True)
top_20.head(20)
```
